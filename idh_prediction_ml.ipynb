```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
import warnings

# Model Seçimi ve Değerlendirme Araçları
from sklearn.model_selection import (
    train_test_split, RepeatedStratifiedKFold,
    RandomizedSearchCV, GridSearchCV
)
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.metrics import (
    precision_score, recall_score, f1_score,
    roc_auc_score, confusion_matrix, make_scorer,
    roc_curve, auc, average_precision_score, accuracy_score,
    matthews_corrcoef, balanced_accuracy_score, precision_recall_curve
)
from sklearn.utils.class_weight import compute_sample_weight

# İMBALANCE ÇÖZÜMÜ İÇİN
from imblearn.over_sampling import SMOTE
from imblearn.pipeline import Pipeline as ImbPipeline

# SEÇİLEN MODELLER
from sklearn.neural_network import MLPClassifier
from xgboost import XGBClassifier

# Görselleştirme ve Uyarı Ayarları
warnings.filterwarnings('ignore')
sns.set_theme(style="whitegrid")

# Karşılaştırma için sonuç sözlüğü
all_results = {}

data_path = "/content/drive/MyDrive/dataset_10.csv"
df = pd.read_csv(data_path, sep=';', decimal='.', index_col=False)

drop_cols = ["D_ConductividadBicarbonato", "D_FlujoBano", "D_Temperatura", "D_PesoPost"]
df.drop(columns=[c for c in drop_cols if c in df.columns], inplace=True)

columns_translation = {
    'D_Hipotension': 'Hypotension_Status',
    'D_Sexo': 'Sex',
    'D_Edad': 'Age',
    'D_Dializador': 'Dialyzer',
    'D_Bano': 'Bath',
    'D_Tecnica': 'Type_of_technique',
    'D_PesoSeco': 'Weight_dry',
    'D_PesoPre': 'Weight_pre',
    'D_Ganancia': 'Interdialytic_weight_gain',
    'D_TemperaturaBano': 'Hemodialysis_bath_temperature',
    'D_VolumenReposicion': 'Volume_changes',
    'D_Kt': 'Urea_clearance',
    'D_FlujoSangre': 'Blood_flow_rate',
    'D_ConductividadBano': 'Hemodialysis_bath_conductivity',
    'D_PresionArterial': 'Arterial_pressure',
    'D_PresionVenosa': 'Venous_pressure',
    'D_PTM': 'Transmembrane_pressure',
    'D_TAS': 'Systolic_Blood_Pressure',
    'D_TAD': 'Diastolic_Blood_Pressure',
    'D_Pulso': 'Heart_rate',
    'D_UF': 'Total_ultrafiltrations'
}

df.rename(columns=columns_translation, inplace=True)
# Yüksek korelasyon (0.99) nedeniyle Weight_dry değişkenini veri setinden çıkarıyoruz
df.drop(columns=['Weight_dry'], inplace=True, errors='ignore')

df['Hypotension_Status'] = df['Hypotension_Status'].map({'SI': 1, 'NO': 0})


# VERİ SETİ TANIMLAYICI VE İSTATİSTİKSEL ANALİZ

print("\n" + "=" * 80)
print("VERİ SETİ GENEL KARAKTERİSTİĞİ VE SINIF DAĞILIMI")
print("=" * 80)

# Veri Seti Boyutu
total_rows = df.shape[0]
total_cols = df.shape[1]
print(f"\n[1] VERİ SETİ BOYUTU:")
print(f"    Toplam Kayıt (Satır)       : {total_rows:,}")
print(f"    Toplam Özellik (Sütun)     : {total_cols}")
print(f"    Toplam Hücre Sayısı        : {total_rows * total_cols:,}")

# Sınıf Dağılımı (Hipotansiyon Durumu)
counts = df["Hypotension_Status"].value_counts()
no_count = counts.get(0, 0)
si_count = counts.get(1, 0)

print(f"\n[2] SINIF DAĞILIMI (HEDEF DEĞİŞKEN):")
print(f"    Hipotansiyon YOK (0 - No)  : {no_count:,} (%{no_count/total_rows*100:.2f})")
print(f"    Hipotansiyon VAR (1 - Yes) : {si_count:,} (%{si_count/total_rows*100:.2f})")
print("-" * 40)

if si_count > 0:
    imbalance_ratio = no_count / si_count
    print(f"    İmbalans Oranı             : {imbalance_ratio:.2f}:1")
else:
    print("Veri setinde hipotansiyon vakası (Sınıf 1) bulunamadı!")

# Eksik Veri ve Veri Kalitesi
missing_total = df.isnull().sum().sum()
print(f"\n[3] EKSİK VERİ VE KALİTE KONTROLÜ:")
if missing_total == 0:
    print("    Durun: Veri setinde hiç eksik hücre bulunmuyor.")
else:
    print(f"    Toplam Eksik Hücre         : {missing_total}")
    print(f"    Eksik Veri Oranı           : %{(missing_total / (total_rows * total_cols)) * 100:.2f}")
    print("\n    Eksik Veri Bulunan Sütunlar:")
    print(df.isnull().sum()[df.isnull().sum() > 0])

# Değişken Tiplerinin Analizi
numerical_features = df.select_dtypes(include=[np.number]).columns.tolist()
if 'Hypotension_Status' in numerical_features:
    numerical_features.remove('Hypotension_Status')
categorical_features = df.select_dtypes(exclude=[np.number]).columns.tolist()

print(f"\n[4] DEĞİŞKEN TİPLERİ:")
print(f"    Sayısal Özellik Sayısı     : {len(numerical_features)}")
print(f"    Kategorik Özellik Sayısı   : {len(categorical_features)}")
if categorical_features:
    print(f"    Kategorik Sütunlar         : {categorical_features}")

print("=" * 80)

print("\n" + "=" * 60)
print("İSTATİSTİKSEL ÖZET")
print("=" * 60)

important_cols = [
    'Age',
    'Weight_pre',
    'Systolic_Blood_Pressure',
    'Diastolic_Blood_Pressure',
    'Heart_rate',
    'Total_ultrafiltrations'
]

existing_cols = [c for c in important_cols if c in df.columns]
print(df[existing_cols].describe().T)

print("\n" + "=" * 60)
print("AYKIRI DEĞER GÖRSELLEŞTİRME (BOXPLOT)")
print("=" * 60)

plt.figure(figsize=(15, 8))
sns.boxplot(data=df[existing_cols], palette="Set2")
plt.xticks(rotation=45)
plt.title("Kritik Klinik Değişkenlerde Aykırı Değer Dağılımı", fontsize=15)
plt.ylabel("Değer Aralığı")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.savefig('00_Boxplot_Outliers.png', dpi=300, bbox_inches='tight')
plt.show()

# GRUP KARŞILAŞTIRMASI
print("\n" + "=" * 80)
print("GRUP KARŞILAŞTIRMASI: HİPOTANSİYON (Yes) vs NORMAL (No)")
print("=" * 80)

from scipy.stats import ttest_ind

# Değişken isimleri (si_df, no_df) aynı kaldı sadece çıktılar değişti
si_df = df[df['Hypotension_Status'] == 1]
no_df = df[df['Hypotension_Status'] == 0]

comparison_cols = ['Age', 'Systolic_Blood_Pressure', 'Diastolic_Blood_Pressure',
                   'Heart_rate', 'Total_ultrafiltrations', 'Weight_pre']

# Başlık satırındaki SI ve NO, Hypotension ve Normal olarak güncellendi
print(f"\n{'Özellik':<30} {'Hypotension (M±SD)':<20} {'Normal (M±SD)':<20} {'p-değer':<15} {'Anlamlı'}")
print("-" * 105)

for col in comparison_cols:
    if col in df.columns:
        si_mean = si_df[col].mean()
        si_std = si_df[col].std()
        no_mean = no_df[col].mean()
        no_std = no_df[col].std()

        t_stat, p_value = ttest_ind(si_df[col].dropna(), no_df[col].dropna())

        if p_value < 0.001:
            significant = "Evet ***"
        elif p_value < 0.01:
            significant = "Evet **"
        elif p_value < 0.05:
            significant = "Evet *"
        else:
            significant = "Hayır"

        # Çıktı formatı
        print(f"{col:<30} {si_mean:6.2f}±{si_std:5.2f}        {no_mean:6.2f}±{no_std:5.2f}        {p_value:<15.4f} {significant}")

print("\n*** p<0.001, ** p<0.01, * p<0.05")


# KATEGORİK KODLAMA
df['Sex'] = df['Sex'].replace({1: 1, 2: 0})
df['Type_of_technique'] = df['Type_of_technique'].replace({1: 1, 2: 0})
df = pd.get_dummies(df, columns=['Dialyzer', 'Bath'], prefix=['Dial', 'Bath'])

print(f"\nToplam Sütun Sayısı: {len(df.columns)}")

# X VE y AYIRMA
X = df.drop(columns=['Hypotension_Status'])
y = df['Hypotension_Status']

print(f"Toplam Özellik Sayısı: {X.shape[1]}")
print(f"Hedef Değişken: {y.name}")

target_col = 'Hypotension_Status'

# KORELASYON ANALİZİ
print("\n" + "=" * 80)
print("KORELASYON ANALIZI")
print("=" * 80)

numerical_df = df.select_dtypes(include=[np.number])
correlation_matrix = numerical_df.corr()

plt.figure(figsize=(14, 12))
mask = np.triu(np.ones_like(correlation_matrix, dtype=bool))

sns.heatmap(
    correlation_matrix,
    mask=mask,
    annot=True,
    fmt='.2f',
    cmap='coolwarm',
    center=0,
    square=True,
    linewidths=1,
    cbar_kws={"shrink": 0.8},
    vmin=-1,
    vmax=1
)
plt.title('Korelasyon Matrisi (Pearson)', fontsize=16, fontweight='bold', pad=20)
plt.tight_layout()
plt.savefig('correlation_matrix_full.png', dpi=300, bbox_inches='tight')
plt.show()

target_corr = correlation_matrix[target_col].drop(target_col).sort_values(ascending=False)

print("\n En Güçlü Pozitif Korelasyonlar:")
print(target_corr.head(10))

print("\n En Güçlü Negatif Korelasyonlar:")
print(target_corr.tail(10))

plt.figure(figsize=(10, 8))
colors = ['#e74c3c' if x > 0 else '#3498db' for x in target_corr.values]
target_corr.plot(kind='barh', color=colors)
plt.title(f'Hedef Değişken ile Korelasyonlar', fontsize=16, fontweight='bold')
plt.xlabel('Korelasyon Katsayısı', fontsize=12)
plt.axvline(x=0, color='black', linestyle='--', linewidth=1)
plt.grid(axis='x', alpha=0.3)
plt.tight_layout()
plt.savefig('target_correlations.png', dpi=300, bbox_inches='tight')
plt.show()

# P-TESTİ
print("\n İstatistiksel Anlamlılık Testi (p-değerleri):")
print("-" * 80)

p_values = {}
for col in numerical_df.columns:
    if col != target_col:
        mask_nan = numerical_df[col].notna() & numerical_df[target_col].notna()
        corr_coef, p_value = stats.pearsonr(
            numerical_df.loc[mask_nan, col],
            numerical_df.loc[mask_nan, target_col]
        )
        p_values[col] = p_value

p_values_df = pd.DataFrame({
    'Korelasyon': target_corr,
    'p-deger': pd.Series(p_values),
    ' Anlamli (p<0.05)': pd.Series(p_values) < 0.05
}).sort_values('Korelasyon', key=abs, ascending=False)

print(p_values_df)

# MULTİCOLLİNEARİTY ANALİZİ
features_only = correlation_matrix.drop(target_col, axis=0).drop(target_col, axis=1)
high_corr_threshold = 0.80
high_correlations = []

for i in range(len(features_only.columns)):
    for j in range(i+1, len(features_only.columns)):
        corr_value = features_only.iloc[i, j]
        if abs(corr_value) > high_corr_threshold:
            high_correlations.append({
                'Feature 1': features_only.columns[i],
                'Feature 2': features_only.columns[j],
                'Correlation': corr_value
            })

if high_correlations:
    print(f"\nUYARI: Yüksek korelasyon tespit edildi (|r| > {high_corr_threshold}):")
    # Listeyi bir DataFrame yaparak şık bir tablo olarak basalım
    high_corr_df = pd.DataFrame(high_correlations)
    print(high_corr_df.to_string(index=False))

    print("\nNOT: Bu değişkenler birbirine çok benzer bilgi taşıyor.")
    print("XGBoost gibi ağaç tabanlı modeller bunu yönetebilir ancak yorumlanabilirliği artırmak için dikkat edilmelidir.")
else:
    print(f"\nONAY: Multicollinearity problemi tespit edilmedi (|r| <= {high_corr_threshold})")

# TRAIN/VAL/TEST SPLIT
X_train_val, X_test, y_train_val, y_test = train_test_split(
    X, y, test_size=0.15, stratify=y, random_state=42
)

X_train, X_val, y_train, y_val = train_test_split(
    X_train_val, y_train_val, test_size=0.18, stratify=y_train_val, random_state=42
)

print(f"\nTrain: {len(X_train)} ({len(X_train)/len(X)*100:.1f}%)")
print(f"Validation: {len(X_val)} ({len(X_val)/len(X)*100:.1f}%)")
print(f"Test: {len(X_test)} ({len(X_test)/len(X)*100:.1f}%)")

# SINIF DENGELEMESİ STRATEJİSİ
print("\n" + "=" * 70)
print("SINIF DENGELEMESİ STRATEJİSİ (ADİL KARŞILAŞTIRMA)")
print("=" * 70)

count_0 = (y_train == 0).sum()
count_1 = (y_train == 1).sum()
imbalance_ratio = count_0 / count_1

print(f"Train Sınıf Dağılımı:")
print(f"  Sınıf 0: {count_0}")
print(f"  Sınıf 1: {count_1}")
print(f"  İmbalance: {imbalance_ratio:.2f}:1")
print(f"\nStrateji: Her iki model için de SMOTE (Sentetik Veri Üretimi)")
print(f"  - XGBoost: SMOTE ile dengelenmiş veri")
print(f"  - MLP: SMOTE ile dengelenmiş veri")
print("=" * 70)

# ÖN İŞLEYİCİ
numerical_cols = X_train.columns.tolist()
preprocessor = ColumnTransformer(
    transformers=[('num', StandardScaler(), numerical_cols)]
)

feature_names = X_train.columns.tolist()
n_features = X_train.shape[1]

# CV STRATEJİSİ
cv_strategy = RepeatedStratifiedKFold(n_splits=5, n_repeats=3, random_state=42)
f1_scorer = make_scorer(f1_score)
recall_scorer = make_scorer(recall_score)

# THRESHOLD OPTİMİZASYONU
def find_best_threshold(y_true, y_prob, metric='f1'):
    thresholds = np.arange(0.05, 0.95, 0.01)
    best_t, best_score = 0.5, 0
    for t in thresholds:
        y_pred = (y_prob >= t).astype(int)
        if metric == 'recall':
            score = recall_score(y_true, y_pred, zero_division=0)
        elif metric == 'f1':
            score = f1_score(y_true, y_pred, zero_division=0)
        if score > best_score:
            best_score = score
            best_t = t
    return best_t, best_score

def calculate_metrics(y_true, y_pred, y_prob):
    tn, fp, fn, tp = confusion_matrix(y_true, y_pred).ravel()
    precisions, recalls, _ = precision_recall_curve(y_true, y_prob)
    pr_auc = auc(recalls, precisions)
    return {
        'accuracy': accuracy_score(y_true, y_pred),
        'precision': precision_score(y_true, y_pred, zero_division=0),
        'recall': recall_score(y_true, y_pred),
        'f1': f1_score(y_true, y_pred),
        'roc_auc': roc_auc_score(y_true, y_prob),
        'pr_auc': pr_auc,
        'avg_precision': average_precision_score(y_true, y_prob),
        'mcc': matthews_corrcoef(y_true, y_pred),
        'balanced_accuracy': balanced_accuracy_score(y_true, y_pred),
        'specificity': tn / (tn + fp) if (tn + fp) > 0 else 0,
        # NPV : "Hipotansiyon yok" tahmininin doğruluğu
        'npv': tn / (tn + fn) if (tn + fn) > 0 else 0
    }

# BOLUM 1: XGBOOST (SMOTE İLE)

print("\n" + "=" * 100)
print("BOLUM 1: XGBOOST (SMOTE İLE)")
print("=" * 100)

print("\nRandomized Search...")
xgb_param_random = {
    'model__n_estimators': [300, 500, 700, 1000],
    'model__max_depth': [4, 6, 8, 10],
    'model__learning_rate': [0.01, 0.03, 0.05, 0.1],
    'model__subsample': [0.7, 0.8, 0.9],
    'model__colsample_bytree': [0.7, 0.8, 0.9],
    'model__gamma': [0, 0.1, 0.2, 0.4],
    'model__reg_alpha': [0, 0.1, 0.5],
    'model__reg_lambda': [1, 1.5, 2],
    'model__min_child_weight': [1, 3, 5]
}

# Pipeline Yapılandırması (SMOTE dahil)
smote_xgb = SMOTE(random_state=42, k_neighbors=5)
xgb_pipe = ImbPipeline([
    ('prep', preprocessor),
    ('smote', smote_xgb),
    ('model', XGBClassifier(eval_metric='logloss', random_state=42, n_jobs=-1))
])

# Rastgele Arama - Eğitim aşamasında genel denge için F1 skorunu kullanıyoruz
xgb_random_search = RandomizedSearchCV(
    xgb_pipe, param_distributions=xgb_param_random, n_iter=80,
    scoring='f1', cv=cv_strategy, random_state=42, n_jobs=-1, verbose=1
)

xgb_random_search.fit(X_train, y_train)
xgb_random_best = xgb_random_search.best_params_
print(f"En iyi CV F1 (RandomSearch): {xgb_random_search.best_score_:.4f}")

print("\nGrid Search...")
xgb_param_grid = {k: [v] for k, v in xgb_random_best.items()}

xgb_grid_search = GridSearchCV(
    xgb_pipe, param_grid=xgb_param_grid, scoring='f1',
    cv=cv_strategy, n_jobs=-1, verbose=0
)

xgb_grid_search.fit(X_train, y_train)
xgb_best_model = xgb_grid_search.best_estimator_
xgb_cv_score = xgb_grid_search.best_score_
print(f"Final CV F1 (GridSearch): {xgb_cv_score:.4f}")

print("\nThreshold (Eşik) Optimizasyonu")
# Doğrulama seti üzerinde olasılıkları hesaplama
xgb_val_prob = xgb_best_model.predict_proba(X_val)[:, 1]
# F1 yaparak dengeyi (Hassasiyet vs Özgüllük) kuruyoruz
xgb_threshold, xgb_val_f1 = find_best_threshold(y_val, xgb_val_prob, metric='f1')
print(f"Optimal Threshold: {xgb_threshold:.3f}, Validation F1: {xgb_val_f1:.4f}")

print("\nÖzellik Önemi Analizi")
xgb_model_obj = xgb_best_model.named_steps['model']
xgb_importances = xgb_model_obj.feature_importances_
xgb_indices = np.argsort(xgb_importances)[::-1]

# Görselleştirme
fig, axes = plt.subplots(1, 2, figsize=(16, 6))
top_n = min(15, len(xgb_indices))
top_indices = xgb_indices[:top_n]

# Sol Grafik: Feature Importance
axes[0].barh(range(top_n), xgb_importances[top_indices], color='#e74c3c', alpha=0.7, edgecolor='black')
axes[0].set_yticks(range(top_n))
axes[0].set_yticklabels([feature_names[i] for i in top_indices], fontsize=9)
axes[0].set_xlabel('Önem Skoru', fontsize=11, fontweight='bold')
axes[0].set_title('XGBoost - En Önemli 15 Özellik', fontsize=12, fontweight='bold')
axes[0].invert_yaxis()
axes[0].grid(axis='x', alpha=0.3)

# Sağ Grafik: Cumulative Importance
cumsum_xgb = np.cumsum(np.sort(xgb_importances)[::-1])
axes[1].plot(range(1, len(cumsum_xgb)+1), cumsum_xgb, 'r-', linewidth=2.5)
axes[1].axhline(y=0.9, color='r', linestyle='--', linewidth=2, label='%90 Eşik')
axes[1].set_xlabel('Özellik Sayısı', fontsize=11, fontweight='bold')
axes[1].set_ylabel('Kümülatif Önem', fontsize=11, fontweight='bold')
axes[1].set_title('XGBoost - Kümülatif Özellik Önemi', fontsize=12, fontweight='bold')
axes[1].legend()
axes[1].grid(alpha=0.3)
plt.tight_layout()
plt.show()

print("\nTest Seti Değerlendirmesi")
xgb_test_prob = xgb_best_model.predict_proba(X_test)[:, 1]
xgb_test_pred = (xgb_test_prob >= xgb_threshold).astype(int)
xgb_test_metrics = calculate_metrics(y_test, xgb_test_pred, xgb_test_prob)

# Karmaşıklık Matrisi Yazdırma
tn, fp, fn, tp = confusion_matrix(y_test, xgb_test_pred).ravel()
print(f"\nKARMASIKLIK MATRISI (XGBoost):")
print(f"True Negatif: {tn}, False Pozitif: {fp}")
print(f"False Negatif: {fn}, True Pozitif: {tp}")

print(f"\nTEST METRİKLERİ:")
for metric, value in xgb_test_metrics.items():
    print(f"   {metric.capitalize():15s}: {value:.4f}")

# Karmaşıklık Matrisi Isı Haritası
cm_xgb = confusion_matrix(y_test, xgb_test_pred)
fig, axes = plt.subplots(1, 2, figsize=(14, 5))
sns.heatmap(cm_xgb, annot=True, fmt='d', cmap='Reds', ax=axes[0],
            cbar_kws={'label': 'Örnek Sayısı'}, annot_kws={'size': 16, 'weight': 'bold'})
axes[0].set_title(f'XGBoost - Confusion Matrix\nRecall: {xgb_test_metrics["recall"]:.3f}', fontsize=12, fontweight='bold')
axes[0].set_ylabel('Gerçek Değer')
axes[0].set_xlabel('Tahmin Edilen')
axes[0].set_xticklabels(['Hipotansiyon Yok', 'Hipotansiyon Var'])
axes[0].set_yticklabels(['Hipotansiyon Yok', 'Hipotansiyon Var'])

# Metrik Bar Grafiği
m_list = ['accuracy', 'precision', 'recall', 'specificity', 'npv', 'f1', 'roc_auc']
m_labels = ['Acc', 'Pre', 'Rec', 'Spec', 'NPV', 'F1', 'AUC']
m_values = [xgb_test_metrics[m] for m in m_list]

# axes[1] üzerinde çizim (Confusion Matrix'in yanındaki grafik)
bars = axes[1].bar(m_labels, m_values, color='#e74c3c', alpha=0.7, edgecolor='black', linewidth=1.5)
axes[1].set_ylim([0, 1.1])
axes[1].set_title('XGBoost - Kritik Klinik Metrikler', fontsize=12, fontweight='bold')
axes[1].grid(axis='y', alpha=0.3)

# Çubukların üzerine değerlerini yazdırma
for bar, val in zip(bars, m_values):
    axes[1].text(bar.get_x() + bar.get_width()/2, val + 0.02, f'{val:.3f}',
                ha='center', va='bottom', fontsize=9, fontweight='bold')

plt.tight_layout()
plt.show()

# Sonuçları Kaydet
all_results['XGBoost'] = {
    'cv_f1': xgb_cv_score,
    'best_params': xgb_grid_search.best_params_,
    'threshold': xgb_threshold,
    'test_metrics': xgb_test_metrics,
    'y_pred': xgb_test_pred,
    'y_prob': xgb_test_prob,
    'color': '#e74c3c'
}

print("\nXGBOOST ANALİZİ TAMAMLANDI!")

print("\nSHAP (Model Açıklanabilirlik) Analizi")
try:
    import shap
    # Veriyi önce transform ediyoruz
    X_test_processed_raw = xgb_best_model.named_steps['prep'].transform(X_test)
    feature_names_out = xgb_best_model.named_steps['prep'].get_feature_names_out()

    # SHAP'in sütun isimlerini doğru okuması için DataFrame'e çeviriyoruz
    X_test_processed = pd.DataFrame(X_test_processed_raw, columns=feature_names_out)

    explainer = shap.TreeExplainer(xgb_best_model.named_steps['model'])
    shap_values = explainer.shap_values(X_test_processed)

    # SHAP değerlerinin boyut kontrolü
    if isinstance(shap_values, list):
        shap_v = shap_values[1]
    elif len(shap_values.shape) == 3:
        shap_v = shap_values[:, :, 1]
    else:
        shap_v = shap_values

    # 1. Grafik: Summary Plot (Değişkenlerin yönü: Pozitif/Negatif etki)

    plt.figure(figsize=(12, 8))
    shap.summary_plot(shap_v, X_test_processed, feature_names=feature_names_out, show=False)
    plt.title('XGBoost - SHAP Öznitelik Etki Analizi (Summary)', fontsize=13, fontweight='bold', pad=15)
    plt.tight_layout()
    plt.savefig('01_XGB_shap_summary.png', dpi=300, bbox_inches='tight')
    plt.show()

    # 2. Grafik: Bar Plot (Değişkenlerin mutlak gücü: En önemli özellikler)

    plt.figure(figsize=(10, 8))
    shap.summary_plot(shap_v, X_test_processed, feature_names=feature_names_out, plot_type="bar", show=False)
    plt.title('XGBoost - Ortalama Mutlak SHAP Değerleri', fontsize=13, fontweight='bold', pad=15)
    plt.tight_layout()
    plt.savefig('01_XGB_shap_bar.png', dpi=300, bbox_inches='tight')
    plt.show()

except Exception as e:
    print(f"SHAP Analizi sırasında bir hata oluştu: {e}")


# BOLUM 2: MLP (SMOTE İLE)

print("\n" + "=" * 100)
print("BOLUM 2: MLP (SMOTE İLE)")
print("=" * 100)

n_features = X_train.shape[1]

print("\nRandomized Search(Dengeli Mod: 40 Iterasyon)")
mlp_param_random = {
    'model__hidden_layer_sizes': [
        (128,), (256,),
        (128, 64), (256, 128),
        (256, 128, 64), # Daha derin ilişkileri yakalamak için
        (n_features,), (n_features*2, n_features)
    ],
    'model__activation': ['relu', 'tanh'],
    'model__solver': ['adam'],
    'model__alpha': [0.001, 0.01, 0.1],
    'model__learning_rate': ['adaptive'],
    'model__learning_rate_init': [0.001, 0.01],
    'model__max_iter': [1500],
    'model__early_stopping': [True],
    'model__validation_fraction': [0.15],
    'model__n_iter_no_change': [20],
    'model__batch_size': ['auto', 128]
}

# SMOTE ile Pipeline Yapılandırması
smote_mlp = SMOTE(random_state=42, k_neighbors=5)
mlp_pipe = ImbPipeline([
    ('prep', preprocessor),
    ('smote', smote_mlp),
    ('model', MLPClassifier(random_state=42))
])

# n_iter=40 ve cv=5 (Toplam 200 fits)
mlp_random_search = RandomizedSearchCV(
    mlp_pipe, param_distributions=mlp_param_random, n_iter=40,
    scoring='f1', cv=5, random_state=42, n_jobs=-1, verbose=1
)

mlp_random_search.fit(X_train, y_train)
mlp_random_best = mlp_random_search.best_params_
print(f"En iyi CV F1 (RandomSearch): {mlp_random_search.best_score_:.4f}")

print("\nGrid Search...")
mlp_param_grid = {k: [v] for k, v in mlp_random_best.items()}
# en stabil noktayı buluyoruz
mlp_param_grid['model__alpha'] = [mlp_random_best['model__alpha'] * 0.9,
                                  mlp_random_best['model__alpha'],
                                  mlp_random_best['model__alpha'] * 1.1]

mlp_grid_search = GridSearchCV(
    mlp_pipe, param_grid=mlp_param_grid, scoring='f1',
    cv=5, n_jobs=-1, verbose=0
)

mlp_grid_search.fit(X_train, y_train)
mlp_best_model = mlp_grid_search.best_estimator_
mlp_cv_score = mlp_grid_search.best_score_
print(f"Final CV F1 (GridSearch): {mlp_cv_score:.4f}")

print("\nThreshold (Eşik) Optimizasyonu - F1 BALANCE")
mlp_val_prob = mlp_best_model.predict_proba(X_val)[:, 1]
# Klinik denge için metric='f1' kullanımı
mlp_threshold, mlp_val_f1 = find_best_threshold(y_val, mlp_val_prob, metric='f1')
print(f"Optimal Threshold: {mlp_threshold:.3f}, Validation F1: {mlp_val_f1:.4f}")

print("\nAğ Yapısı Bilgileri")
mlp_model_obj = mlp_best_model.named_steps['model']
print(f"Gizli Katman Yapısı: {mlp_model_obj.hidden_layer_sizes}")
print(f"Aktivasyon Fonksiyonu: {mlp_model_obj.activation}")
print(f"Eğitimdeki Toplam İterasyon: {mlp_model_obj.n_iter_}")

print("\nTest Seti Değerlendirmesi")
mlp_test_prob = mlp_best_model.predict_proba(X_test)[:, 1]
mlp_test_pred = (mlp_test_prob >= mlp_threshold).astype(int)
mlp_test_metrics = calculate_metrics(y_test, mlp_test_pred, mlp_test_prob)

tn_mlp, fp_mlp, fn_mlp, tp_mlp = confusion_matrix(y_test, mlp_test_pred).ravel()
print(f"\nKARMASIKLIK MATRİSİ (MLP):")
print(f"True Negative: {tn_mlp}, False Positive: {fp_mlp}")
print(f"False Negative: {fn_mlp}, True Positive: {tp_mlp}")

print(f"\nTEST METRİKLERİ:")
for metric, value in mlp_test_metrics.items():
    print(f"   {metric.capitalize():15s}: {value:.4f}")

# Karmaşıklık Matrisi Isı Haritası
cm_mlp = confusion_matrix(y_test, mlp_test_pred)
fig, axes = plt.subplots(1, 2, figsize=(14, 5))
sns.heatmap(cm_mlp, annot=True, fmt='d', cmap='Blues', ax=axes[0],
            cbar_kws={'label': 'Örnek Sayısı'}, annot_kws={'size': 16, 'weight': 'bold'})
axes[0].set_title(f'MLP - Confusion Matrix\nRecall: {mlp_test_metrics["recall"]:.3f}',
                 fontsize=12, fontweight='bold', pad=15)
axes[0].set_ylabel('Gerçek Değer')
axes[0].set_xlabel('Tahmin Edilen')
axes[0].set_xticklabels(['No', 'Yes'])
axes[0].set_yticklabels(['No', 'Yes'])

# Metrik Bar Grafiği
m_names = ['accuracy', 'precision', 'recall', 'specificity', 'npv', 'f1', 'roc_auc']
m_labels = ['Acc', 'Pre', 'Rec', 'Spec', 'NPV', 'F1', 'AUC']
m_vals = [mlp_test_metrics[m] for m in m_names]

# axes[1] üzerinde çizim (Confusion Matrix'in yanındaki grafik)
bars = axes[1].bar(m_labels, m_vals, color='#3498db', alpha=0.7, edgecolor='black', linewidth=1.5)
axes[1].set_ylim([0, 1.1])
axes[1].set_title('MLP - Kritik Klinik Metrikler', fontsize=12, fontweight='bold')
axes[1].grid(axis='y', alpha=0.3)

# Çubukların üzerine değerlerini yazdırma
for bar, val in zip(bars, m_vals):
    axes[1].text(bar.get_x() + bar.get_width()/2, val + 0.02, f'{val:.3f}',
                ha='center', va='bottom', fontsize=9, fontweight='bold')

plt.tight_layout()
plt.savefig('04_MLP_confusion_matrix.png', dpi=300, bbox_inches='tight')
plt.show()

# Sonuçları Kaydet
all_results['MLP'] = {
    'cv_f1': mlp_cv_score,
    'best_params': mlp_grid_search.best_params_,
    'threshold': mlp_threshold,
    'test_metrics': mlp_test_metrics,
    'y_pred': mlp_test_pred,
    'y_prob': mlp_test_prob,
    'color': '#3498db'
}

print("\nMLP ANALİZİ TAMAMLANDI!")

print("\nSHAP (Açıklanabilirlik) Analizi")
try:
    import shap
    mlp_model_obj = mlp_best_model.named_steps['model']

    # Veriyi işleyip DataFrame'e çeviriyoruz (Sütun isimleri için)
    X_train_raw = mlp_best_model.named_steps['prep'].transform(X_train)
    X_test_raw = mlp_best_model.named_steps['prep'].transform(X_test)
    feature_names_out = mlp_best_model.named_steps['prep'].get_feature_names_out()

    X_train_df = pd.DataFrame(X_train_raw, columns=feature_names_out)
    X_test_df = pd.DataFrame(X_test_raw, columns=feature_names_out)

    # Örnekleme 100 ÖRNEK
    background_data = shap.sample(X_train_df, 100)
    test_samples = shap.sample(X_test_df, 100)

    # Explainer kurulumu
    explainer_mlp = shap.KernelExplainer(lambda x: mlp_model_obj.predict_proba(x)[:, 1], background_data, link="identity")

    shap_values_mlp = explainer_mlp.shap_values(test_samples)

    # 1. Grafik: Summary Plot (Noktalı Dağılım - Etki Yönü)
    plt.figure(figsize=(12, 8))
    shap.summary_plot(shap_values_mlp, test_samples, feature_names=feature_names_out, show=False)
    plt.title('MLP - SHAP Öznitelik Etki Analizi (Summary)', fontsize=13, fontweight='bold', pad=15)
    plt.tight_layout()
    plt.savefig('02_MLP_shap_summary.png', dpi=300, bbox_inches='tight')
    plt.show()

    # 2. Grafik: Bar Plot (Çubuk Grafik - Önem Derecesi)
    plt.figure(figsize=(10, 8))
    shap.summary_plot(shap_values_mlp, test_samples, feature_names=feature_names_out, plot_type="bar", show=False)
    plt.title('MLP - Ortalama Mutlak SHAP Değerleri (Bar Plot)', fontsize=13, fontweight='bold', pad=15)
    plt.tight_layout()
    plt.savefig('02_MLP_shap_bar.png', dpi=300, bbox_inches='tight')
    plt.show()

    print("   MLP SHAP analizi başarıyla tamamlandı!")
except Exception as e:
    print(f"   MLP SHAP Analizi sırasında hata: {e}")

# BOLUM 3: FINAL KARSILASTIRMA VE RAPORLAMA

print("\n" + "=" * 100)
print("BOLUM 3: FINAL KARSILASTIRMA")
print("=" * 100)

comparison_data = []

for model_name, results in all_results.items():
    m = results['test_metrics']
    comparison_data.append({
        'Model': model_name,
        'Accuracy': m['accuracy'],
        'Precision': m['precision'],
        'Recall (Sens.)': m['recall'], # Duyarlılık
        'Specificity': m['specificity'], # Özgüllük
        'NPV': m['npv'], # Negatif Tahmin Değeri
        'F1-Score': m['f1'],
        'ROC-AUC': m['roc_auc'],
        'MCC': m['mcc'],
        'Threshold': results['threshold']
    })

comparison_df = pd.DataFrame(comparison_data).set_index('Model')
print("\n" + comparison_df.round(4).to_string())
comparison_df.to_csv('FINAL_XGB_MLP_detailed_comparison.csv')
print("\nCSV kaydedildi: FINAL_XGB_MLP_detailed_comparison.csv")

# ROC Eğrileri
print("\nROC egrileri")
plt.figure(figsize=(10, 8))
for model_name, results in all_results.items():
    fpr, tpr, _ = roc_curve(y_test, results['y_prob'])
    roc_auc_val = auc(fpr, tpr)
    plt.plot(fpr, tpr, lw=3, label=f'{model_name} (AUC={roc_auc_val:.3f})',
            color=results['color'])

plt.plot([0, 1], [0, 1], 'k--', lw=2, label='Random (AUC=0.50)')
plt.xlim([0, 1])
plt.ylim([0, 1.05])
plt.xlabel('False Positive Rate (1-Specificity)', fontsize=12, fontweight='bold')
plt.ylabel('True Positive Rate (Recall)', fontsize=12, fontweight='bold')
plt.title('ROC Curves - Model Comparison (SMOTE-Based)', fontsize=15, fontweight='bold', pad=15)
plt.legend(loc="lower right", fontsize=10)
plt.grid(alpha=0.3)
plt.tight_layout()
plt.savefig('05_FINAL_roc_curves.png', dpi=300, bbox_inches='tight')
plt.show()

# Precision-Recall Eğrileri
print("\nPrecision-Recall egrileri")
plt.figure(figsize=(10, 8))
for model_name, results in all_results.items():
    precision_curve, recall_curve, _ = precision_recall_curve(y_test, results['y_prob'])
    avg_prec = average_precision_score(y_test, results['y_prob'])
    plt.plot(recall_curve, precision_curve, lw=3,
             label=f'{model_name} (AP={avg_prec:.3f})',
             color=results['color'])

baseline = (y_test == 1).sum() / len(y_test)
plt.axhline(y=baseline, color='k', linestyle='--', lw=2,
            label=f'Baseline (Prevalans)={baseline:.3f}')
plt.xlim([0, 1])
plt.ylim([0, 1.05])
plt.xlabel('Recall (Duyarlılık)', fontsize=12, fontweight='bold')
plt.ylabel('Precision (Kesinlik)', fontsize=12, fontweight='bold')
plt.title('Precision-Recall Curves', fontsize=15, fontweight='bold', pad=15)
plt.legend(loc="lower left", fontsize=10)
plt.grid(alpha=0.3)
plt.tight_layout()
plt.savefig('06_FINAL_precision_recall.png', dpi=300, bbox_inches='tight')
plt.show()

# Metrik Karşılaştırma Grafiği
print("\nMetrik karsilastirmasi")
metrics_list = ['Accuracy', 'Precision', 'Recall (Sens.)', 'Specificity', 'NPV', 'F1-Score', 'ROC-AUC']

fig, axes = plt.subplots(2, 4, figsize=(20, 10))
axes = axes.ravel()

for idx, metric in enumerate(metrics_list):
    model_names = list(all_results.keys())
    values = [comparison_df.loc[name, metric] for name in model_names]
    colors = [all_results[name]['color'] for name in model_names]

    bars = axes[idx].bar(range(len(model_names)), values, color=colors, alpha=0.8, edgecolor='black', linewidth=1.5)
    axes[idx].set_xticks(range(len(model_names)))
    axes[idx].set_xticklabels(model_names, rotation=15, ha='right', fontsize=10)
    axes[idx].set_ylabel(metric, fontsize=11, fontweight='bold')
    axes[idx].set_title(f'{metric} Comparison', fontsize=12, fontweight='bold')
    axes[idx].set_ylim([0, 1.1])
    axes[idx].grid(axis='y', alpha=0.3)

    for bar, val in zip(bars, values):
        axes[idx].text(bar.get_x() + bar.get_width()/2, val + 0.02, f'{val:.3f}',
                      ha='center', va='bottom', fontsize=10, fontweight='bold')

# Boş kalan 8. kutu gizlendi
if len(axes) > len(metrics_list):
    axes[7].axis('off')

plt.tight_layout()
plt.savefig('07_FINAL_metrics_comparison.png', dpi=300, bbox_inches='tight')
plt.show()

# ÖZET RAPOR

best_recall_model = comparison_df['Recall (Sens.)'].idxmax()
best_auc_model = comparison_df['ROC-AUC'].idxmax()

print("\n" + "=" * 50)
print("KLİNİK ÖZET RAPOR")
print("-" * 50)
print(f"Hipotansiyon Vakalarını En İyi Yakalayan (Recall): {best_recall_model}")
print(f"Genel Ayırt Etme Gücü En Yüksek Olan (AUC): {best_auc_model}")
print(f"Negatif Tahmin Güvenirliği (NPV) Ortalama: {comparison_df['NPV'].mean():.4f}")
print("=" * 50)

print("\nÖZET RAPOR")
print("=" * 100)
best_recall = max(all_results.items(), key=lambda x: x[1]['test_metrics']['recall'])
best_auc = max(all_results.items(), key=lambda x: x[1]['test_metrics']['roc_auc'])

print(f"\nEN YUKSEK RECALL: {best_recall[0]}")
print(f"   Recall: {best_recall[1]['test_metrics']['recall']:.4f}")
print(f"   AUC: {best_recall[1]['test_metrics']['roc_auc']:.4f}")

print(f"\nEN YUKSEK AUC: {best_auc[0]}")
print(f"   AUC: {best_auc[1]['test_metrics']['roc_auc']:.4f}")
print(f"   Recall: {best_auc[1]['test_metrics']['recall']:.4f}")

print("\n" + "=" * 100)
print("ANALIZ TAMAMLANDI")
print("=" * 100)
```
